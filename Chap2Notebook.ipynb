# Stratify Sampling

### `median_income` is very important to predict housing prices

```python
housing['median_income'].hist()
```

### Most median income are clustered from 1.5 to 6. - Sub categorize the data into 5 categories (labeled 1 to 5): category range from 0 to 1.5, and 1.5 to 3...)

```python
housing['income_cat']=pd.cut(housing['median_income'], bins=[0,1.5,3,4.5,6,np.inf],
                            labels=[1,2,3,4,5])
#cat=category
```

```python
housing['income_cat'].hist()
```

```python
from sklearn.model_selection import train_test_split
strat_training_set, strat_test_set=train_test_split(housing, test_size=0.2, stratify=housing['income_cat'],
                                                   random_state=42)
```

```python
strat_training_set.info()
```

```python
# remove income_cat
for cat in (strat_training_set, strat_test_set):
    cat.drop("income_cat", axis=1, inplace=True)
```

```python
strat_training_set.info()
```

```python
housing_new=strat_training_set.copy()
```

```python
housing_new.info()
```

# Data Cleaning

### Imputing missing values

```python
housing_new.info()
```

```python
housing_new['total_bedrooms'].hist()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median') # can select most_frequent
```

```python
housing_new['total_bedrooms']=imputer.fit_transform
```

```python
housing_new.info()
```

### One hot shot encoding using dummy variable

```python
housing_new.columns
```

```python
housing_new=pd.get_dummies(housing[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value', 'ocean_proximity']], drop_first=True)
```
